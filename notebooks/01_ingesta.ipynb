```python
import duckdb
import pandas as pd

# Conectamos (o creamos) la base de datos local
con = duckdb.connect("lakehouse.duckdb")

# Cargamos los CSV crudos tal cual, sin transformar nada
df_customers_raw = pd.read_csv("../data/raw/customers.csv")
df_products_raw = pd.read_csv("../data/raw/products.csv")
df_orders_raw = pd.read_csv("../data/raw/orders.csv")

# Los registramos como tablas RAW en DuckDB
con.execute("CREATE OR REPLACE TABLE raw_customers AS SELECT * FROM df_customers_raw")
con.execute("CREATE OR REPLACE TABLE raw_products AS SELECT * FROM df_products_raw")
con.execute("CREATE OR REPLACE TABLE raw_orders AS SELECT * FROM df_orders_raw")

print("Tablas RAW creadas correctamente.")
con.execute("SHOW TABLES").df()
```

```python
con.close()
```

## Resumen - Ingesta de Datos

Se cargaron los 3 archivos CSV (customers, products, orders) usando Pandas, 
y se almacenaron como tablas RAW en una base de datos local DuckDB 
(`lakehouse.duckdb`), sin aplicar ninguna transformación, preservando 
los datos exactamente como llegaron.

Tablas creadas: raw_customers, raw_products, raw_orders.
